```python
import pandas as pd

# Define aggre_transaction as a Pandas DataFrame
aggre_transaction = pd.DataFrame({
    'States': ['State1', 'State2', 'State3'],
    'Transaction_count': [100, 200, 300]
})

# Group the data by states and sum the transaction counts
state_transaction_counts = aggre_transaction.groupby('States')['Transaction_count'].sum().reset_index()

# Create the histogram
fig = px.histogram(state_transaction_counts, x='Transaction_count', nbins=31, title='Distribution of Transaction Counts per State')
fig.show()
```

```python

aggre_transaction['Quarter'] = [1, 2, 3]
quarter_transaction_counts = aggre_transaction.groupby('Quarter')['Transaction_count'].sum().reset_index()
fig = px.bar(quarter_transaction_counts, x='Quarter', y='Transaction_count', title='Transaction Counts by Quarter')
fig.update_xaxes(title='Quarter')
fig.update_yaxes(title='Transaction Count')
fig.show()

```
